```python
import pandas as pd
```

```python
# 减少评论
origin=pd.read_csv('comments_original.csv')
```

```python
movies=pd.read_csv('movie-2.2.csv')
```

```python
asins=set()
for i,r in movies.iterrows():
    for a in r['Asin'].split(','):
        asins.add(a)
```

```python
asins
```

```python
comments=origin[origin['productId'].map(lambda a:a in asins)]
```

```python
comments
```

```python
comments = comments.reset_index()
```

```python
comments['review_id']=comments.index
```

```python
comments
```

```python
from datetime import datetime
datetime.fromtimestamp(1180828800).strftime("%Y-%m-%d %H:%M:%S")
```

```python
for i,r in comments.iterrows():
    comments.at[i,'time']=datetime.fromtimestamp(int(r['time'])).strftime("%Y-%m-%d %H:%M:%S")
```

```python
comments
```

```python
comments.to_csv('comments-1.1.csv',index=False)
```
