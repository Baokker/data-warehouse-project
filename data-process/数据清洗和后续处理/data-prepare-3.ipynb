```python
import pandas as pd
movies=pd.read_csv('movie-2.2.csv')
```

```python
movies
```

```python
Asin=pd.DataFrame(columns=['asin','movie_id'])
```

```python
import gc

count=0
for i,r in movies.iterrows():
    for f in str(r['Asin']).split(','):
        Asin=pd.concat([Asin,pd.DataFrame(data={'asin':[f],'movie_id':[r['ID']]})],ignore_index=True)

    count+=1
    if count%1000==0:
        print(count,end=' ')
        gc.collect()
```

```python
Asin
```

```python
Asin.to_csv('sql-data/Asin.csv',index=False)
```
