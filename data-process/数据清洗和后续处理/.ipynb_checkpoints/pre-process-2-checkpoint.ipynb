```python
# 先提取movie id
movie_id=set()
count=0
for index,row in comments.iterrows():
    count+=1
    if count%10000==0:
        print(count,end=' ')
    if 'movie' in row['text'] or 'film' in row['text']:
        movie_id.add(str(row['productId']))
```

```python
len(movie_id)
```

```python
movie_id_df=pd.DataFrame(data={'id':list(movie_id)})
```

```python
movie_id_df.to_csv('movie_id.csv',index=False,sep=',')
```

```python
def filter_movie(asin_str:str):
    for a in asin_str.split(','):
        if a in movie_id:
            return True
    return False
```

```python
new_movies = movies[movies['Asin'].map(filter_movie)]
```
