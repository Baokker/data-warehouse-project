```python
import pandas as pd
```

```python
movies=pd.read_csv('movie-2.2.csv')
```

```python
movies
```

```python
Genre=pd.DataFrame(columns=['genre_id','genre_name','movie_id','movie_title'])
```

```python
import gc

count=0
for i,r in movies.iterrows():
    if not pd.isna(r['Genres']):
        Genre=pd.concat([Genre,pd.DataFrame(data={'genre_name':[r['Genres']],'movie_id':[r['ID']],'movie_title':[r['Title']]})],ignore_index=True)
        
    count+=1
    if count%1000==0:
        print(count,end=' ')
        gc.collect()
```

```python
Genre
```

```python
Genre['genre_id']=Genre.index
```

```python
Genre
```

```python
Genre.to_csv('sql-data/Genre.csv',index=False)
```
