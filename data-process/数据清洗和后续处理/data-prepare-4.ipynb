```python
import pandas as pd
```

```python
movies=pd.read_csv('sql-data/Movie.csv')
actors=pd.read_csv('sql-data/Actor.csv')
directors=pd.read_csv('sql-data/Director.csv')
info=pd.read_csv('movie-2.2.csv')
```

```python
Act=pd.DataFrame(columns=['movie_id','movie_title','actor_id'])
```

```python
actor_dict={}
for i,r in actors.iterrows():
    actor_dict[r['name']]=r['actor_id']
```

```python
import gc

collections=[]
count=0
for i,r in info.iterrows():
    if not pd.isna(r['Actors']):
        for a in r['Actors'].split(','):
            collections.append(pd.DataFrame(data={'actor_id':[actor_dict[a]],\
                                              'movie_id':[r['ID']],'movie_title':[r['Title']]}))
    count+=1
    if count%1000==0:
        print(count,end=' ')
        gc.collect()

# 最后再concat 否则复杂度是N^2！
Act=pd.concat(collections,ignore_index=True)
```

```python
Act
```

```python
Act.to_csv('sql-data/Act.csv',index=False)
```

```python
Cooperation=pd.DataFrame(columns=['movie_id','type','left_person_id','right_person_id'])
```

```python
import gc
collections=[]
count=0
for i,r in info.iterrows():
    if not pd.isna(r['Directors']) and not pd.isna(r['Actors']):
        for d in r['Directors'].split(','):
            for a in r['Actors'].split(','):
                collections.append(pd.DataFrame(data={'left_person_id':[director_dict[d]],\
                                                      'right_person_id':[actor_dict[a]],\
                                                      'movie_id':[r['ID']],'type':[1]}))
    
    if not pd.isna(r['Actors']):
        actor_list=list(map(lambda a:actor_dict[a],r['Actors'].split(',')))
        actor_list.sort()
        for m in range(len(actor_list)):
            for n in range(m+1,len(actor_list)):
                collections.append(pd.DataFrame(data={'left_person_id':[actor_list[m]],\
                                                      'right_person_id':[actor_list[n]],\
                                                      'movie_id':[r['ID']],'type':[2]}))
    count+=1
    if count%1000==0:
        print(count,end=' ')
        gc.collect()

# 最后再concat 否则复杂度是N^2！
Cooperation=pd.concat(collections,ignore_index=True)
```

```python
Cooperation
```

```python
Cooperation.to_csv('sql-data/Cooperation.csv',index=False)
```
