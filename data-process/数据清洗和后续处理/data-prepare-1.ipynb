```python
# 导出演员 导演
import gc
from datetime import datetime

actors=set()
directors=set()
genres=set()
formats=set()
times=set()

count=0
for i,r in movies.iterrows():
    count+=1
    if count%1000==0:
        print(count,end=' ')
        gc.collect()
    
    if not pd.isna(r['Actors']):
        for a in str(r['Actors']).split(','):
            actors.add(a)
            
    if not pd.isna(r['Directors']):
        for d in str(r['Directors']).split(','):
            directors.add(d)
            
    if not pd.isna(r['Genres']):
        for g in str(r['Genres']).split(','):
            genres.add(g)
            
    if not pd.isna(r['Format']):
        for f in str(r['Format']).split(','):
            genres.add(f)
    
#     if not pd.isna(r['Release date']):
#         times.add(datetime.strptime(str(r['Release date']),"%B %d, %Y").strftime("%Y-%m-%d %H:%M:%S"))
```

```python
Actor=pd.DataFrame(data={'name':list(actors)})
Actor['actor_id']=Actor.index
Actor.to_csv('sql-data/Actor.csv',index=False)

Director=pd.DataFrame(data={'name':list(directors)})
Director['director_id']=Director.index
Director.to_csv('sql-data/Director.csv',index=False)
```

```python
import warnings
warnings.filterwarnings('ignore') # setting ignore as a parameter

import gc

collections=[]
count=0
for i,r in movies.iterrows():
    if not pd.isna(r['Format']):
        for f in str(r['Format']).split(','):
    #         Format = Format.append({ 'format_name':f,'movie_id':r['ID'],'movie_title':r['Title']},ignore_index=True)
#             Format=pd.concat([Format,pd.DataFrame(data={'format_name':[f],'movie_id':[r['ID']]})],ignore_index=True)
            collections.append(pd.DataFrame(data={'format_name':[f],'movie_id':[r['ID']],'movie_title':r['Title']}))
    count+=1
    if count%1000==0:
        print(count,end=' ')
        gc.collect()

# 最后再concat 否则复杂度是N^2！
Format=pd.concat(collections,ignore_index=True)
```

```python
from datetime import datetime

collections=[]
count=0
for i,r in movies.iterrows():
    if not pd.isna(r['Release date']):
        date=datetime.strptime(r['Release date'],"%B %d, %Y")
        collections.append(pd.DataFrame(data={'date':[date.strftime("%Y-%m-%d %H:%M:%S")],\
                                              'movie_id':[r['ID']],'year':[date.year],\
                                              'month':[date.month],'day':[date.day],\
                                              'season':[date.month//4+1],'weekday':[date.isoweekday()]}))
    count+=1
    if count%1000==0:
        print(count,end=' ')
        gc.collect()

# 最后再concat 否则复杂度是N^2！
ReleaseDate=pd.concat(collections,ignore_index=True)
```

```python
ReleaseDate['time_id']=ReleaseDate.index
```
