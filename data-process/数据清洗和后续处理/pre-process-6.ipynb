```python
import pandas as pd
```

```python
movies=pd.read_csv('movie-2.1.csv')
```

```python
# 增加format
movies['Format']=''
```

```python
movies
```

```python
origin=pd.read_csv('movies_info_full_date.csv')
```

```python
origin
```

```python
import gc

count=0
for index,row in movies.iterrows():
    count+=1
    if count % 1000 ==0:
        print(count,end=' ')
        gc.collect()
    
    asins=str(row['Asin']).split(',')
    movie_format=set()
    for i,r in origin[origin['ASIN'].map(lambda t:t in asins)].iterrows():
        if not pd.isna(r['Media Format']):
            for t in str(r['Media Format']).split(','):
                movie_format.add(t)

    movies.at[index,'Format']=','.join(movie_format)
```

```python
movies['ID']=movies.index
```

```python
movies.to_csv('movie-2.2.csv',index=False)
```
