```python
import os
import gc

i=0
for index,row in movies.iterrows():
    i+=1
    if i%1000 ==0:
        print(i,end=' ')
        # 内存太大了...
        gc.collect()
    
    if pd.isna(row['Genres']):
        for a in str(row['Asin']).split(','):
            file_path='WebPages//'+a+'.html'
            if os.path.exists(file_path):
                html = etree.parse(file_path,etree.HTMLParser())
                r = html.xpath('//*[@id="wayfinding-breadcrumbs_feature_div"]/ul/li/span/a/text()')
                
                if len(r)<1:
                    continue
                else:
                    genre=r[-1].strip()
                    if genre != '':
                        movies.at[index,'Genres']=genre
                        break
```
