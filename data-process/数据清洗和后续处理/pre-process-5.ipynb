```python
core_comments = comments.drop(columns=['userId', 'profileName', 'helpfulness','time', 'summary', 'text'])
```

```python
core_comments
```

```python
comments_score = core_comments.groupby(['productId'],as_index=False).agg({'score':'mean'})
```

```python
# import warnings
# warnings.filterwarnings('ignore') # setting ignore as a parameter
import gc
from statistics import mean
count=0

# 增加版本和评分
for index,row in movies.iterrows():
    count+=1
#     print(count,end=' ')
    if count%1000 == 0:
        print(count,end=' ')
        gc.collect()
    
#     scores=[]
#     for i,r in pd.concat([core_comments[core_comments['productId']==a] for a in str(row['Asin']).split(',')]).iterrows():
#         scores.append(float(str(r['score'])))
    
    if row['Score'] == 0:
        scores=[]
        asins=str(row['Asin']).split(',')
        for i,r in comments_score[comments_score['productId'].map(lambda t: t in asins)].iterrows():
            scores.append(float(str(r['score'])))

        movies.at[index,'Score']=mean(scores)
        movies.at[index,'Edition']=len(asins)
```

```python
movies.to_csv('movie-2.1.csv',index=False,sep=',')
```
