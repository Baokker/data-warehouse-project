```python
import pandas as pd
```

```python
movie_info=pd.read_csv('movies_info_final.csv')
movie_asin=pd.read_csv('merged_title_movies_source_asin_collection.csv')
```

```python
# 移除asin表中的多余元素
movie_asin
```

```python
movie_asin.columns.values
```

```python
movie_asin.drop(['Language', 'Release date', 'Date First Available',
       'Run time', 'Producers', 'Directors', 'Writers', 'Actors',
       'Media Format', 'Subtitles', 'Genres', 'Writer'],axis=1,inplace=True)
movie_asin.rename(columns={'  Title':'Asin'},inplace=True)
```

```python
movie_asin
```

```python
# 合并
movies=pd.concat([movie_info,movie_asin],axis=1)
```

```python
movies
```

```python
movies.columns.values
```

```python
# 去除多余的列
movies.drop(['Language', 'Date First Available','Run time', 'Producers','Writers','Media Format', 'Subtitles',],axis=1,inplace=True)
```

```python
movies
```

```python
# 修缮asin和actors格式
import string

i=0
bad_actor=['- Various','*','VARIOUS','various','Various']
for index, row in movies.iterrows():
    i+=1
    if i%1000 == 0:
        print(i,end=' ')
        
    actors=str(row['Actors']).strip("[]").split(', ')
    actors=[s.replace("'","").replace('"','').strip() for s in actors]
    actors=[string.capwords(s) for s in actors if string.capwords(s) not in bad_actor]
    movies.at[index,'Actors']=','.join(actors)
    
    asins=str(row['Asin']).strip("[]").split(', ')
    asins=[s.replace("'","").replace('"','').strip() for s in asins]
    movies.at[index,'Asin']=','.join(asins)
```

```python
movies
```

```python
movies.to_csv('movie-1.1.csv',index=False,sep=',')
```
