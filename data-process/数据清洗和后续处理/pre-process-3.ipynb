```python
movies=pd.read_csv('movie-1.5.csv')
movies.sort_values(by=['Title'], ignore_index=True,inplace=True)
```

```python
merged_movies = pd.DataFrame(columns=['Title', 'Release date', 'Directors', 'Actors', 'Genres', 'Asin'])
```

```python
','.join([a for a in set(''.split(',')+'A,B,C'.split(',')) if a != ''])
```

```python
import datetime

i = 0

whole_length = movies.shape[0]

for index, row in movies.iterrows():
    i += 1
    if whole_length - index <= 30: # 不能直接用shape[0]，因为会不断变化
        end = whole_length
    else:
        end = index + 30
        
    similar_title_info = process.extract(row['Title'], movies.loc[index:end, 'Title'] , limit = 30)
    similar_title_order = [s[2] for s in similar_title_info if s[1] >= 95]        
    
    # WARNING:root:Applied processor reduces input query to empty string, all comparisons will have score 0. [Query: '-']
    # '$','-'
    if similar_title_order == [] and index in movies.index:
        similar_title_order = [index]
    # has been added
    elif similar_title_order == []:
        continue
    
    for_merge_info = movies.loc[similar_title_order,:]
    
    merged_movie_info = pd.DataFrame(data=[['','','','','','']],columns=['Title', 'Release date', 'Directors', 'Actors', 'Genres', 'Asin'])
    
    # 初始
    for index, row in for_merge_info.iterrows():
        # Title 选最短名
        if merged_movie_info.loc[0,'Title'] == '' or len(merged_movie_info.loc[0,'Title']) > len(row['Title']): 
            merged_movie_info.loc[0,'Title'] = row['Title']
        # Release Date选最早的
        if not pd.isna(row['Release date']):
            if merged_movie_info.loc[0,'Release date'] == '':
                merged_movie_info.loc[0,'Release date'] = row['Release date']
            else:
                merged_date = datetime.datetime.strptime(merged_movie_info.loc[0,'Release date'],"%B %d, %Y")
                cur_date = datetime.datetime.strptime(row['Release date'],"%B %d, %Y")
                if merged_date > cur_date:
                    merged_movie_info.loc[0,'Release date'] = row['Release date']
                    
        # Directors 
        if not pd.isna(row['Directors']):
            merged_movie_info.loc[0,'Directors'] = ','.join([a for a in set(str(row['Directors']).split(',')+str(merged_movie_info.loc[0,'Directors']).split(',')) if a!=''])

        # Actors
        if not pd.isna(row['Actors']):
            merged_movie_info.loc[0,'Actors'] = ','.join([a for a in set(str(row['Actors']).split(',')+str(merged_movie_info.loc[0,'Actors']).split(',')) if a!=''])
      
        # Genres 选第一个出现的
        if not pd.isna(row['Genres']) and merged_movie_info.loc[0,'Genres'] == '':
            merged_movie_info.loc[0,'Genres'] = row['Genres']
            
        # Asin
        if not pd.isna(row['Asin']):
            merged_movie_info.loc[0,'Asin'] = ','.join([a for a in set(str(row['Asin']).split(',')+str(merged_movie_info.loc[0,'Asin']).split(',')) if a!=''])
    
    merged_movies=pd.concat([merged_movies,merged_movie_info], ignore_index=True)    
    movies.drop(similar_title_order, inplace=True)
    
    if i % 1000 == 0:
        print(i)
```

```python
merged_movies
```

```python
merged_movies.to_csv('movie-1.7.csv',index=False,sep=',')
```
